# Data Cleaning

```python
# Removing the '%' of the end of international and domestic columns and change them to float
# str: works with every cell of the series as a string
domestic=pixar_movies['Domestic %'].str.rstrip('%').astype(float)
international=pixar_movies['International %'].str.rstrip('%').astype(float)
```

```python
if pixar_movies["Domestic %"].dtypes=='object' and pixar_movies['International %'].dtypes=='object':
    pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
    pixar_movies['International %']=pixar_movies['International %'].str.rstrip('%').astype(float)
pixar_movies.dtypes
```

```python
#Create a new DataFrame named filtered_pixar with the last row filtered since it has some missing data
filtered_pixar=pixar_movies.dropna()
filtered_pixar
```

** Reindexing the datasets with the name of the Movies**

```python
pixar_movies.set_index(pixar_movies['Movie'],inplace=True)
filtered_pixar.set_index(filtered_pixar['Movie'],inplace=True)
```
