```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

forest_fires=pd.read_csv("forestfires.csv")
forest_fires.head(5)
```

```python
sns.distplot(forest_fires['temp'], kde=False)
sns.plt.show()
```

```python
plt.hist(forest_fires['temp'],bins=50)
sns.plt.show()
```

As we mentioned before, there are 2 primary ways to customize how a Seaborn plot looks:

* Plotting function parameters, e.g. sns.distplot(kde=False)
* Seaborn functions, called on the sns Seaborn object synonym

Similar to how you use plt, Matplotlib's import synonym, to customize Matplotlib graphics, you use sns, Seaborn's
import synonym, to refer to a plot and call functions from Seaborn's API to customize it. To set the x-axis label, we pass the xlabel keyword argument into the distplot method.

```python
sns.distplot(forest_fires['wind'], kde=False, axlabel='wind speed')
sns.plt.show()
```

```python
# Using Syle
sns.set_style('dark')
sns.distplot(forest_fires['wind'],kde=False, axlabel='wind speed')
sns.plt.show()
```

# Box Plots
